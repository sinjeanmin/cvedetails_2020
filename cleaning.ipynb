```python
import pandas as pd
from datetime import datetime as dt
```

```python
df = pd.read_feather("https://github.com/sinjeanmin/cvedetails_2020_final/raw/main/cve_2020_final_16.feather")
```

```python
#casting all cvss_score to integer, from string
for index in df.index:
    cvss_score = df.loc[index, "cvss_score"]
    cvss_score = pd.to_numeric(cvss_score)
    df.loc[index, "cvss_score"] = cvss_score
```

```python
df['response_time'] = None
```

```python
#adding column response_time, and the running operation earliest_patch minus publish_date
for index in df.index:
    has_patch = df.loc[index, "has_patch"]
    if has_patch == 1:
        publish_date = dt.strptime(df.loc[index, "publish_date"], "%Y-%m-%d")
        earliest_patch = dt.strptime(df.loc[index, "earliest_patch"], "%Y-%m-%d")
        response_time = earliest_patch - publish_date
        response_time = response_time.days
        df.loc[index, "response_time"] = response_time
```

```python
df['complexity_low']= [1 if x=='Low' else 0 for x in df['complexity']]
df['complexity_med']= [1 if x=='Medium' else 0 for x in df['complexity']]
df['complexity_high']= [1 if x=='High' else 0 for x in df['complexity']]
```

```python
network = 0
adj_network = 0
for index in df.index:
    access_level = df.loc[index, "access_level"]
    if access_level == "Remote":
        df.loc[index, "access_level"] = "Network"
        network = network + 1
    elif access_level == "Local Network":
        df.loc[index, "access_level"] = "Adjacent Network"
        adj_network = adj_network + 1
```

```python
df['access_network']= [1 if x=='Network' else 0 for x in df['access_level']]
df['access_local']= [1 if x=='Local' else 0 for x in df['access_level']]
df['access_adjnetwork']= [1 if x=='Adjacent Network' else 0 for x in df['access_level']]
```

```python
df['product_app']= [1 if x=='Application' else 0 for x in df['product_type']]
df['product_os']= [1 if x=='OS' else 0 for x in df['product_type']]
df['product_hardware']= [1 if x=='Hardware' else 0 for x in df['product_type']]
```

```python
df['cvss_low'] = None
df['cvss_medium'] = None
df['cvss_high'] = None

for index in df.index:
    cvss_score = df.loc[index,'cvss_score']
    if cvss_score <= 3.9:
        df.loc[index, 'cvss_low'] = 1
        df.loc[index, 'cvss_medium'] = 0
        df.loc[index, 'cvss_high'] = 0
    elif cvss_score > 3.9 and cvss_score <= 6.9:
        df.loc[index, 'cvss_low'] = 0
        df.loc[index, 'cvss_medium'] = 1
        df.loc[index, 'cvss_high'] = 0
    elif cvss_score > 6.9 and cvss_score <= 10:
        df.loc[index, 'cvss_low'] = 0
        df.loc[index, 'cvss_medium'] = 0
        df.loc[index, 'cvss_high'] = 1
```

```python
df['cvss_bin']=pd.cut(df.cvss_score,bins=[0.0,3.9,6.9,10],labels=['Low','Medium','High'])
```
